```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

```python
print("A.A−1 = I:")
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print(identity_check)
```

```python
print("AB ≠ BA:")
AB = np.dot(A, B)
BA = np.dot(B, A)
print("AB:")
print(AB)
print("BA:")
print(BA)
print("Are they equal?", np.array_equal(AB, BA))
```

```python
print("(AB)T = BT AT:")
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)
print("(AB)T:")
print(AB_T)
print("BT AT:")
print(BT_AT)
print("Are they equal?", np.array_equal(AB_T, BT_AT))
```

```python
print("Solve system of linear equations:")
coefficients = np.array([[2, -3, 1],
                         [1, -1, 2],
                         [3, 1, -1]])
constants = np.array([-1, -3, 9])

inv_coefficients = np.linalg.inv(coefficients)
solution = np.dot(inv_coefficients, constants)
print(solution)
```

```python
import time
import numpy as np

size = 1000000
```

```python
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
list_time = time.time() - start_time
print(f"Python list time: {list_time:.6f} seconds")
```

```python
arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()
result_arr = arr1 + arr2
numpy_time = time.time() - start_time
print(f"NumPy array time: {numpy_time:.6f} seconds")
```
